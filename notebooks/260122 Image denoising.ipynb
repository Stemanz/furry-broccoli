```python
from furrybroccoli import Denoiser
from PIL import Image
import keras
```

```python
!ls
```

```python
img = Image.open("reallife 256x192.jpg")
img
```

```python
d = Denoiser(
    img,
    keras.models.load_model("model2_ds2_27x512_strides2_13x256_strides2_RMSProp_mse"),
    tile_size=64
)
```

### Simple denoising. We will get tiling artefacts

```python
d.denoise()
```

### Advanced denoising. We will get a clear image at the cost of processing time

```python
d.adv_denoise()
```
